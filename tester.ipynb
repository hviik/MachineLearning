```python
T = np.linspace(0, 10, 50)
np.random.seed(0)

# Toy derivative of cost wrt w over a number of iterations
dw = np.exp(-0.3*T) * np.random.uniform(low=0, high=2, size=(50,))

# First moment (velocity)
q1 = 0.9
v = np.zeros((50,))
for t in range(1, 50): 
    v[t] = (q1 * v[t-1]) + ((1-q1) * dw[t]) / (1 - (q1**t))

# Second moment (squared gradient)
q2 = 0.999
s = np.zeros((50,))
for t in range(1, 50): 
    s[t] = (q2 * s[t-1]) + ((1-q2) * (dw[t]**2)) / (1 - (q2**t))

# Adam computation with small epsilon
e = 1e-9
adam = v / np.sqrt(s + e)
```

```python
# Create plot
plt.figure(figsize=(8, 6))
plt.style.use('dark_background')  # Set background to black

# Plot the curves with different colors
plt.plot(T, dw, color='lightpink', label='dw')     # Red for dw
plt.plot(T, v, color='lightgreen', label='v')     # Green for v
plt.plot(T, adam, color='lightblue', label='adam') # Blue for adam

# Add labels and title
plt.xlabel('T')
plt.ylabel('Values')
plt.title('Plot of dw, v, and adam')

# Add legend to identify colors
plt.legend()

# Display plot
plt.show()
```

```python
def softmax (Z_): 
    expZ_ = np.exp(Z_)
    return expZ_/(np.sum(expZ_, axis = 0, keepdims = True))

def crossENT (A, Y): 
    L = -np.sum(Y * np.log(A), axis = 0, keepdims = True)
    m = A.shape[1]
    J = np.sum(L)/m
    return J 

def prelu (H, p): 
    return np.where(H > 0, H, p * H)
def prelu_grad (H,p):
    return np.where(H > 0, 1, p) 

def fprop (A_prev, W, B, V, C, p): 
    Z = np.dot(W, A_prev) + B
    mean = np.mean(Z, axis = 1, keepdims = True)
    sig = np.std(Z, axis = 1, keepdims = True)  + 0.000001
    Z_ = (Z - mean)/sig
    H = (V * Z_) + C
    A = prelu (H,p) 

    fcache = {'A_prev': A_prev, 'Z': Z, 'Z_': Z_, 'H': H, 'A': A}
    params = {'W': W, 'B': B, 'V': V, 'C': C}
    return fcache, params

def bprop (fcache, params, dA, p, m): 
    A_prev = fcache['A_prev']
    Z = fcache['Z']
    Z_ = fcache['Z_']
    H = fcache['H']
    A = fcache['A']

    W = params['W']
    B = params['B']
    V = params['V']
    C = params['C']
    dH = dA * prelu_grad(H,p)
    dV = np.sum((Z_ * dH), axis = 1, keepdims = True)
    dC = np.sum(dH, axis = 1, keepdims = True) 
    dZ_ = V * dH 
    sig = np.std(Z, axis = 1, keepdims = True) + 0.000001
    dZ = ((m*dZ_) - (np.sum(dZ_, axis = 1, keepdims = True)) - (Z_ * (np.sum((Z_ * dZ_), axis = 1, keepdims = True))))/(m * sig)
    dW = np.dot(dZ, A_prev.T)
    dB = np.sum(dZ, axis = 1, keepdims = True)
    dA_prev = np.dot(W.T, dZ)
    grads = {'dV' : dV, 'dC' : dC, 'dW' : dW, 'dB' : dB, 'dA_prev' : dA_prev}
    return grads

def fin_fprop (A_prev, W, B, V, C, p): 
    Z = np.dot(W, A_prev) + B
    mean = np.mean(Z, axis = 1, keepdims = True)
    sig = np.std(Z, axis = 1, keepdims = True)  + 0.000001
    Z_ = (Z - mean)/sig
    H = (V * Z_) + C
    A = softmax(H)

    fcache = {'A_prev': A_prev, 'Z': Z, 'Z_': Z_, 'H': H, 'A': A}
    params = {'W': W, 'B': B, 'V': V, 'C': C}
    return fcache, params

def fin_bprop (fcache, params, Y, p, m): 
    A_prev = fcache['A_prev']
    Z = fcache['Z']
    Z_ = fcache['Z_']
    H = fcache['H']
    A = fcache['A']

    W = params['W']
    B = params['B']
    V = params['V']
    C = params['C']
    dH = (A - Y)/m
    dV = np.sum((Z_ * dH), axis = 1, keepdims = True)
    dC = np.sum(dH, axis = 1, keepdims = True) 
    dZ_ = V * dH 
    sig = np.std(Z, axis = 1, keepdims = True) + 0.000001
    dZ = ((m*dZ_) - (np.sum(dZ_, axis = 1, keepdims = True)) - (Z_ * (np.sum((Z_ * dZ_), axis = 1, keepdims = True))))/(m * sig)
    dW = np.dot(dZ, A_prev.T)
    dB = np.sum(dZ, axis = 1, keepdims = True)
    dA_prev = np.dot(W.T, dZ)
    grads = {'dV' : dV, 'dC' : dC, 'dW' : dW, 'dB' : dB, 'dA_prev' : dA_prev}
    return grads
    
```
